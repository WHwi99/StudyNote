## Tensor 생성

```python
# 정수형 int64 기본
# 생성 시 dtype 설정하여 실수형으로 변경
t1 = torch.tensor([1,2,3], dtype=torch.float)
print(t1)
print(type(t1))
print(t1.dtype)
print(t1.shape)
print(t1.size())
print(t1.dim())     # 차원
print(t1.device)    # CPU 적재
```

```python
t2 = torch.tensor([[1., 2., 3.], [4., 5., 6.]])
print(t2)
print(type(t2))
print(t2.dtype)
print(t2.shape)
print(t2.size())
print(t2.dim())     # 차원
print(t2.device)    # CPU 적재
```

```python
# 0 또는 1 tensor 생성
print(torch.ones(2, 3))     # 자료형 float32
print(torch.zeros(2, 3, dtype=torch.int))

# 존재하는 tensor의 shape를 기반으로 생성
x = torch.tensor([[1, 2, 3], [4, 5, 6]])
print(torch.ones_like(x))
print(torch.zeros_like(x))
```

```python
# 난수 생성
# - 정규 분포
print(torch.normal(10, 1, (2, 2)))      # mean, std, size
print(torch.randn((2, 2)))              # size
print(torch.randn_like(x.float()))      # float로 변환 후 처리

# - 균등 분포
print(torch.rand((2, 2)))               # size, 0~1 사이 값
print(torch.randint(1, 10, (2, 2)))     # start ~ stop
print(torch.rand_like(x.float()))       # 0~1 사이 값

# - 그 외
print(torch.empty((2, 3)))
```

```python
# numpy ndarray 호환
import numpy as np

x = np.array([1, 2, 3])
print(x, type(x))

t = torch.from_numpy(x)
print(t, type(t))

x2 = t.numpy()
print(x2, type(x2))
```

## 연산

```python
a = torch.tensor([[1, 2, 3], [4, 5, 6]], dtype=torch.float)
b = torch.tensor([[7, 8, 9], [10, 11, 12]], dtype=torch.float)
c = torch.tensor([[1, 2], [3, 4], [5, 6]], dtype=torch.float)

print(a.shape, b.shape, c.shape)
```

```python
# 원소별 연산
print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a // b)
print(a % b)
```

```python
# broadcasting 연산
print(a + 1)
print(a + torch.tensor([10, 20, 30]))
# print(a + torch.tensor([10, 20])) << 행렬의 사이즈가 맞지 않음
```

```python
# 내적 (모든 원소간의 곱을 더하기)
# - 1차원 : torch.dot(), torch.matmul(), @
# - 2차원 : torch.matmul(), @
m = torch.tensor([1, 2, 3])
n = torch.tensor([4, 5, 6])
print(m.dot(n))
print(torch.dot(m, n))
print(torch.matmul(m, n))
print(m @ n)
```

```python
# 2차원 이상의 내적
print(a.matmul(c))      # (2, 3) @ (3, 2) = (2, 2)
print(a @ c)
# print(torch.dot(a, c))  >>  1차원 배열만 사용 가능한 함수 (2차원에서 사용 x)
```

## 통계 연산

```python
# 합계
print(torch.sum(a, dim=0))
print(torch.sum(a, dim=1))
print(torch.sum(a, dim=None))
```

```python
# 평균
print(torch.mean(a, dim=0))
print(torch.mean(a, dim=1))
print(torch.mean(a, dim=None))
```

## 형태 변경

```python
t = torch.arange(12)
print(t)

t_ = t.reshape(3, 4)
t_ = t.reshape(3, -1)
t_ = t.reshape(-1, 4)
t_ = t.reshape(2, 2, 3)

print(t_, t_.size())
```

```python
# 전치
t = torch.arange(12).reshape(3, 4)
print(t, t.shape)
print(t.T, t.T.shape)
print(t.transpose(0,1), t.transpose(0,1).shape)   # 0차원, 1차원 교환
```

```python
# 평탄화
print(t.flatten())
print(t.reshape(-1))
```

```python
# 차원 축소
x = torch.tensor([[[1, 2], [3, 4]]])
print(x.size())

print(x.squeeze(0), x.squeeze(0).shape)
```

```python
# 차원 추가
x = torch.tensor([[1, 2], [3, 4]])
print(x.size())

x_ = x.unsqueeze(0)

print(x_, x_.size())
```

## indexing | slicing

```python
x = torch.arange(1, 13).reshape(3, 2, 2)
x
```

```python
# 인덱싱
print(x[0])
print(x[1, 0])
print(x[2, 1])
print(x[-1, -1])
print(x[-1, 0, -1])
```

```python
# 슬라이싱
x_ = x[:, 0]       # 인덱싱 결과는 2차원
x_ = x[:, :1]      # 슬라이싱 결과는 3차원

print(x_, x_.shape)
```

```python
# fancy indexing
index = [1, 2]
x_ = x[index]

print(x_, x_.shape)
```

## 유용한 메소드

```python
# split
x = torch.zeros(10,4)
print(x)

x_splits = torch.split(x, 4)

for x_ in x_splits:
    print(x_)
```

```python
# expand
k = torch.tensor([[[1, 2]], [[3, 4]]])

k_exp = k.expand(2, 3, 2)
print(k_exp, k_exp.shape)
```

```python
# 원본 배열 k의 view 형태로 k_exp가 생성 되어 k_exp가 k를 참조하고 있다.
k_exp[0, 0, 0] = 100
print(k)
print(k_exp)
```

```python
# repeat
k = torch.tensor([[[1, 2]], [[3, 4]]])

# 반복할 차원의 횟수
k_rep = k.repeat(1, 3, 1)
print(k_rep, k_rep.shape)

k_rep[0, 0, 0] = 100
print(k)
print(k_rep)
```

```python
# masked_fill : mask 조건식이 true 경우만 값 변경
x = torch.arange(9).reshape(3, -1)
print(x)

mask = x > 4
print(mask)

x_ = x.masked_fill(mask, value=0)
print(x_)
```

## 모델 학습에서 tensor 다차원 배열 연산

```python
# 단층 선형 연산
X = torch.randn(4, 5)   # 입력 값 : 4개의 sample, 5개의 feature
print(f'{X = }')

W = torch.randn(5, 3)       # 가중치 (3개의 노드)
print(f'{W = }')
b = torch.randn(1, 3)       # 노드마다 편향(bias)
print(f'{b = }')

# Linear 층 계산
z = torch.matmul(X, W) + b

z
```
