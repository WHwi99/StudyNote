# 주석(Comment)
**주석** : 컴퓨터가 해석하지 않고 무시하는 구문으로 코드 설명 또는 특정 코드의 실행을 방지할 때 사용한다.

```python
# 한 줄 주석
print('이 문장은 실행함')
print('이 문장은 실행 안함')
```

# 변수(Variable) 
데이터를 저장하는 **메모리 공간**의 이름

## 변수 사용 이유
1. 유지보수
2. 재사용
3. 가독성

```python
# '='연산자 : 오른쪽 값을 왼쪽 변수에 대입(할당)
team_name = "ohgiraffers"
print(team_name)

# 값 변경 : 기존 데이터 위에 새로운 값이 덮어 씌워짐
team_name = "오지라퍼스"
print(team_name)

# 값 복사 : 변수에 담긴 값을 다른 변수에 할당 가능
my_team = team_name
print(my_team)
```

## 변수 이름 짓기 (Naming Rule)
### 1) 필수 규칙 (위반하면 에러가 발생)

        - 특수문자/공백 불가
        - 예약어 사용 불가
### 2) 권장 관례 (가독성을 위한 약속)
        - snake_case => 단어와 단어 사이에 공백 대신에 언더바로 연결하는 것

        - 한글 사용 가능하지만 인코딩/호환성 이슈가 발생할 수 있어 지양
