```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 연속형 자료인 성적을 범주형으로 변경
bins = [0, 50, 70, 80, 90, 100]
labels = ['F', 'D', 'C', 'B', 'A']

df['grade'] = pd.cut(df['exam_score'], bins=bins, labels=labels)
df
```

```python
# 성별(범주형)과 생존여부(범주형)의 관계
sns.countplot(data=df, x='sleep_quality', hue='grade')
plt.show()
sns.countplot(data=df, x='gender', hue='grade')
plt.show()
sns.countplot(data=df, x='sleep_quality', hue='grade')
plt.show()
sns.countplot(data=df, x='sleep_quality', hue='grade')
plt.show()
```

```python
# 숫자형 / 범주형 나누기
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns

print("📌 수치형 변수:", num_cols)
print("📌 범주형 변수:", cat_cols)

plt.figure(figsize=(10, 8))
sns.heatmap(df[num_cols].corr(), annot=True, fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
