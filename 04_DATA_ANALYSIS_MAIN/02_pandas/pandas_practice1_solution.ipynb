# pandas_practice1_solution

```python
import pandas as pd
import numpy as np

orders = pd.DataFrame({
    '주문번호': ['O001', 'O002', 'O003', 'O004', 'O005', 'O006', 'O006', 'O008'],
    '주문일자': ['2026-01-03', '2026-01-05', '2026-01-05', '2026-01-08', '2026-01-10', '2026-01-12', '2026-01-12', '2026-01-15'],
    '고객명': ['김철수', '이영희', '박민수', '최지영', '정다은', '홍길동', '홍길동', '강민호'],
    '지역': ['서울', '부산', '서울', '대전', None, '부산', '부산', '서울'],
    '카테고리': ['전자', '식품', '전자', '패션', '식품', '전자', '전자', '패션'],
    '상품명': ['노트북', '사과', '마우스', '자켓', '우유', '키보드', '키보드', '운동화'],
    '수량': [1, 10, 2, 1, np.nan, 1, 1, 2],
    '단가': [1500000, 3000, 25000, 120000, 2500, 50000, 50000, 90000],
    '결제상태': ['완료', '완료', '완료', '취소', '완료', '완료', '완료', '완료']
})
orders
```

## 1. 기본 확인
1) orders의 shape를 확인하시오.
2) columns를 확인하시오.
3) 상위 3행을 확인하시오.

```python
print(orders.shape)
print(orders.columns)
display(orders.head(3))
```

## 2. 열 선택
1) 고객명 열만 선택하시오.
2) 상품명, 수량, 단가 열만 선택하시오.

```python
print(orders['고객명'])
display(orders[['상품명', '수량', '단가']])
```

## 3. 인덱싱
주문번호를 인덱스로 설정한 새로운 DataFrame df를 만들고 아래를 수행하시오.
1) O003 행 전체를 loc으로 조회
2) 첫 번째 행을 iloc으로 조회
3) O002의 단가를 at으로 조회

```python
df = orders.set_index('주문번호')
print(df.loc['O003'])
print(df.iloc[0])
print(df.at['O002', '단가'])
```

## 4. 조건 필터링
1) 결제상태가 완료인 행만 조회하시오.
2) 카테고리가 전자이고 단가가 30000 이상인 행만 조회하시오.
3) 지역이 서울 또는 부산인 행만 조회하시오.

```python
display(orders[orders['결제상태'] == '완료'])
display(orders[(orders['카테고리'] == '전자') & (orders['단가'] >= 30000)])
display(orders[orders['지역'].isin(['서울', '부산'])])
```

## 5. 값 수정
1) 결제상태가 취소인 행의 수량을 0으로 수정하시오.
2) 지역이 결측치인 행의 지역을 '미상'으로 수정하시오.

```python
orders.loc[orders['결제상태'] == '취소', '수량'] = 0
orders.loc[orders['지역'].isna(), '지역'] = '미상'
orders
```

## 6. 파생 열 추가
총액 열을 추가하시오.
총액 = 수량 * 단가

```python
orders['총액'] = orders['수량'] * orders['단가']
orders
```

## 7. 입출력
1) 현재 DataFrame을 orders_raw.csv로 저장하시오. 인덱스는 저장하지 마시오.
2) 저장한 파일을 다시 읽어 orders_csv 변수에 담고 상위 2행을 확인하시오.

```python
orders.to_csv('orders_raw.csv', index=False, encoding='utf-8-sig')
orders_csv = pd.read_csv('orders_raw.csv')
orders_csv.head(2)
```

## 8. 결측값 확인
1) 열별 결측값 개수를 확인하시오.
2) 결측값이 하나라도 있는 행만 조회하시오.

```python
print(orders.isna().sum())
orders[orders.isna().any(axis=1)]
```

## 9. 결측값 처리
1) 수량 결측치는 1로 채우시오.
2) 지역 결측치는 '미상'으로 채우시오.
처리 결과를 orders_clean 변수에 저장하시오.

```python
orders_clean = orders.fillna({'수량': 1, '지역': '미상'})
orders_clean
```

## 10. 중복값 처리
1) 중복 행 개수를 확인하시오.
2) 중복 행을 제거한 결과를 orders_unique 변수에 저장하시오.

```python
orders_clean.duplicated().sum()
orders_unique = orders_clean.drop_duplicates()
orders_unique
```

## 11. 정렬
orders_unique를 기준으로
1) 단가 내림차순 정렬
2) 카테고리 오름차순, 단가 내림차순 정렬
을 수행하시오.

```python
orders_unique.sort_values('단가', ascending=False)
orders_unique.sort_values(['카테고리', '단가'], ascending=[True, False])
```

## 12. 타입 변환
orders_unique의 주문일자 열을 datetime으로 변환하시오.

```python
orders_unique['주문일자'] = pd.to_datetime(orders_unique['주문일자'])
orders_unique.dtypes
```

## 13. 최종 확인
전처리가 끝난 orders_unique에서 필요한 열만 골라 final_orders 변수에 저장하시오.
선택 열: 주문일자, 고객명, 지역, 카테고리, 상품명, 수량, 단가, 총액, 결제상태

```python
final_orders = orders_unique[
    ['주문일자', '고객명', '지역', '카테고리', '상품명', '수량', '단가', '총액', '결제상태']
]
final_orders
```
