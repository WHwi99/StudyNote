# 상속
 - 부모 클래스의 속성과 메소드를 자식 클래스가 물려받아 재사용하고 확장하는 개념

```python
# 상속 기본 문법
class Person:
    national = "korea"
    
    def greeting(self):
        return "Hello. This is Python."

# Person 클래스를 상속하는 Student 클래스
class Student(Person):
    pass    # 문법 오류 없이 코드 실행 시 pass 할 수 있도록


student = Student()
print(f"국적 확인 : {Student.national}")
print(f"메소드 호출 : {student.greeting()}")
```

## 부모 생성자 호출 - super()
 - 자식 클래스에서 새로운 속성을 추가할 때 부모가 가진 기존 속성 초기화를 위해 `super().__init()__`을 호출한다

```python
class Person:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    
    # 인스턴스를 참조하는 변수 출력 시 반환하는 문자열 지정하는 메소드
    def __str__(self):
        return f'<Person {self.name} {self.phone}>'
    
class Employee(Person):
    def __init__(self, name, phone, position, salary):
        # 부모 생성자 호출하여 이름, 전화번호 초기화 위임
        super().__init__(name, phone) # 부모 생성자의 속성을 호출한다.
        self.position = position
        self.salary = salary
    
    def __str__(self):
        # 부모의 __str__결과에 자신의 정보를 추가
        return f'{super().__str__()} - {self.position} {self.salary}'

p1 = Person('홍길동', '010-1234-5678')
print(p1)

e1 = Employee('유관순', '010-9876-4321','대리',3_000_000)
print(e1)
```

## 다중 상속
 - 파이썬은 여러 부모로부터 동시에 상속 받는 것을 지원한다.
 - 탐색 순서는 **왼쪽에서 오른쪽** 순서이다. (MRO: Method Resolution Order)

```python
class Person:
    def greeting(self):
        return 'Hello. This is Python.'
    
class Learner:
    def greeting(self):
        return 'Hello. I am learner.'
    def learn(self):
        return 'I am learning Python.'

# 다중 상속 => 두 개의 클래스를 모두 상속하는 Student 클래스    
class Student(Person, Learner):
    pass

stu = Student()
print(stu.greeting())   # 탐색 순서가 왼쪽에서 오른쪽이기 때문에 같은 이름의 함수가 있다면 먼저 상속한 Person의 greeting()함수가 적용된다.
print(stu.learn())

print(Student.__mro__) # 메소드 탐색 순서를 확인
```
