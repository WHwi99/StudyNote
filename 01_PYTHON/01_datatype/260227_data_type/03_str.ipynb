# 문자열(String)
글자들의 모음. 큰따옴표("") 또는 작은따옴표('')를 사용한다.

```python
# 생성 및 여러 줄 문자열
name = "홍길동"
# 멀티라인 생성
poem = """ 
나 보기가 역겨워
가실 때에는
죽어도 아니 눈물 흘리오리다
"""
print(poem)

# 문자열 연산
greeting = "안녕"
print(greeting + " " + name) # 이어 붙이기
print(name * 3) # 곱한만큼 반복

# [주의] 숫자와 직접 더하는 것은 불가능하다
# print(name + 20) => TypeError: can only concatenate str (not "int") to str
print(name + str(20))
```

# 주요 문자열 메소드
문자열을 다루는 유용한 기능들이 있다. (단, 문자열 원본은 변경되지 않는다.)

```python
# 치환 및 공백 제거
text = "    Hello Python    "
print(text.strip()) # strip()함수를 이용해서 양쪽에 있는 공백을 제거 할 수 있다.
print(text.replace("Python","World")) # replace()를 이용해서 문자열의 특정 부분을 교체할 수 있다.

# 대소문자 변환
s = "PyThOn"
print(s.upper())
print(s.lower())

# 분리와 결합
fruits = "사과, 포도, 바나나"
fruits_list = fruits.split(", ") # split() 특정 문자 기준으로 자르기
print(fruits_list)

joined_fruits = "-".join(fruits_list) # join() 리스트를 하나로 합치기
print(joined_fruits)
```

# 내가 필요한 기능을 어떻게 찾을까?
파이썬의 모든 자료형은 각자 쓸 수 있는 메소드를 가지고 있다.
구글링도 좋지만 코드로 직접 확인하는 버을 알아보자.

```python
simple_str = "Python"

# dir 함수 : 이 자료형이 가진 모든 기능 리스트로 보여줌
print(dir(simple_str))

# help 함수 : 특정 기능의 사용법을 알려줌
help(simple_str.upper)

# IDE의 자동 완성
# 변수명 뒤에 마침표를 찍고 확인
# simple_str. 까지 입력시 자동완성 기능으로 사용 가능한 부분이 뜨게 되는데 현재는 자동 완성 기능을 빼놓았기 때문에 뜨지 않는다.
```

# 문자열 포맷팅
문자열 안에 변수 값을 삽입하는 방법

```python
name = "유관순"
age = 20

# 과거 방식
print("이름 : %s, 나이 : %d" %(name, age))
print("이름 : {0}, 나이 : {1}".format(name,age))

# 권장 방식(f-string)
print(f"이름 : {name}, 나이 : {age}")
```

```python
a = 10
b = 20
print(f"{a} + {b} = {a+b}") # f-string {}내 에서도 간단한 연산도 가능하다.
```
