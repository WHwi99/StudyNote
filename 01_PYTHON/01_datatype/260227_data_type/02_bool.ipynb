# 논리형 (Boolean)
참(True) 또는 거짓(False)을 표현하는 자료형이다.

```python
# 선언 : 첫 글자는 반드시 대문자로 작성하여 따옴표를 사용하지 않음
bool1 = True
bool2 = False

print(bool1, type(bool1))
print(bool2, type(bool2))
```

# 비교 연산
두 값을 비교하여 결과로 논리형을 반환한다

```python
# 숫자 비교
print("10 == 10 : ", 10 == 10)
print("10 != 10 : ", 10 != 10)
print("10 < 5 : ", 10 < 5 )
print("10 > 5 : ", 10 > 5 )
```

# 연쇄 비교와 문자열 사전 순 비교

```python
# 연쇄 비교
age = 25
print("18 <= age <= 30 :", 18 <= age <= 30)

# 문자열 비교 : 사전 순 비교
print("apple < banana :", "apple" < "banana")
print("Apple < apple :", "Apple" < "apple") #대문자가 소문자보다 아스키코드 값이 작음

# [주의] 문자열 vs 숫자 비교
print("10 < 9 :", "10" < "9") # 숫자값을 문자열로 불러오면 반환되는 논리형값이 달라질 수 있다.
print("10 < 9 :", 10 < 9)
```
