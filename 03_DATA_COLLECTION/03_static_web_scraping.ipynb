# 정적 웹 페이지 스크래핑

## Web Crawlig vs Web Scraping
 - Web Crawling : 여러 웹 페이지를 자동으로 순회하며 페이지를 수집하는 작업
 - Web Scraping : 웹 페이지의 HTML을 분석하여 필요한 데이터만 추출하는 작업
  
## 정적 웹 페이지 vs 동적 웹 페이지
 - 정적 웹 페이지(SSR : Server Side Rendering)
 - 동적 웹 페이지(CSR : Client Side Rendering)

## 웹 서비스 흐름
 1. 클라이언트 http://example.com 요청
 2. 웹 서버 HTML 문서 응답
 3. 클라이언트 HTML parsing 후 화면 rendering

## 샘플 페이지를 통해 BeautifulSoup 메소드 동작 확인

### find | findALL

```python
with open('sample.html','r', encoding = 'utf-8') as f:
    html = f.read()
    # html 태그를 계층 구조대로 분석해서 가지고 있음
    soup = BeautifulSoup(html, 'html.parser')
    
    # find_all : 모든 태그 검색
    li_tags = soup.find_all('li')
    for li_tag in li_tags:
        print(li_tag.text)
        
    # find : 첫번째 태그 검색
    li_tag = soup.find("li")
    print(li_tag)
    
    # soup 객체 뿐만 아니라 반환 된 객체를 대상으로 다시 하위 조회
    a_tag = li_tag.find('a')
    print(a_tag)
    
    # 속성 조회
    url = a_tag.attrs['href'] # dict 형태로 속성 값 저장 되어 있음
    url = a_tag.get('href')
    print(url)
    
    # 속성을 포함한 조회
    h1_tag = soup.find('h1', {'id':'page-title'})
    print(h1_tag)
    
    # 모든 .section-content 태그 조회
    section_content_tags = soup.find_all("section", {"class" : "section-content"})
    print(section_content_tags)
```

## select | select_one
 - CSS 선택자를 사용해 태그 조회
 - 태그 : 태그명
 - 아이디 : #아이디명
 - 클래스 : .클래스명

```python
li_tags = soup.select("li")
print(li_tags)

first_li = soup.select_one("li")
print(first_li)

h1_page_title_tag = soup.select_one("h1#page-title")
print(h1_page_title_tag)

section_tags = soup.select("section.section-content")
print(section_tags)
```

## HTML 계층 구조를 이용한 조회
 - 부모 : parent
 - 자식 : find | find_all | select | select_one
 - 형제 : find_previous_sibling | find_next_sibling

```python
p_tags = soup.select("p")
print(p_tags)

# 부모 탐색
print(p_tags[0].parent)

# 자식 탐색
p_children = p_tags[0].find_all()
print(p_children)

# 형제 탐색
print(p_tags[0].find_previous_sibling())
print(p_tags[0].find_next_sibling())
```

## 네이버 뉴스 정적 페이지 웹 스크래핑

```python
import requests
from bs4 import BeautifulSoup as bs

query = '인공지능'
url = f'https://search.naver.com/search.naver?ssc=tab.news.all&where=news&sm=tab_jum&query={query}'

response = requests.get(url)
soup = bs(response.text, "html.parser")

# 뉴스 타이틀 태그 조회
title_tags = soup.select("span.sds-comps-text.sds-comps-text-ellipsis.sds-comps-text-ellipsis-1.sds-comps-text-type-headline1")
titles = [span.text for span in title_tags]

# 뉴스 설명 조회
description_tags = soup.select('span.sds-comps-text.sds-comps-text-ellipsis.sds-comps-text-ellipsis-3.sds-comps-text-type-body1')
descriptions = [span.text for span in description_tags]

# 링크 설명
a_tags = soup.select('div.sds-comps-vertical-layout.sds-comps-full-layout > a:first-of-type')
links = [a.attrs['href'] for a in a_tags]
```

```python
from dataclasses import dataclass
from datetime import datetime
from typing import Optional

@dataclass
class NaverNews:
    
    id : int
    title : str
    originallink : str
    link : str
    description : str
    pubDate : str
    create_at : Optional[datetime] = None
```

```python
news_lines = zip(titles, descriptions, links)
naver_news_list = []

for news in news_lines:
    title, description, link = news
    naver_news_list.append(NaverNews(
        None, title=title, description=description, link=link, originallink=link, pubDate=None
        ))
    
naver_news_list
```

## 이미지 다운로드

```python
import os
from urllib.request import urlretrieve

parent_dir = "./images"
if not os.path.isdir(parent_dir):
    os.mkdir(parent_dir)
    
img_tags = soup.select("div.sds-comps-base-layout.sds-comps-inline-layout.sds-comps-image.fit-contain.forced-ratio img")
img_src_list = [img_tag.attrs['src'] for img_tag in img_tags]

for i, image_src in enumerate(img_src_list):
    urlretrieve(image_src, os.path.join(parent_dir, f"naver_news_{i}.jpg"))
```
