# 람다 (Lambda)
 - 람다 : 이름 없는 한 줄 짜리 함수. 별도의 함수로 선언하기에 간단한 작업을 즉석에서 처리할 때 사용

```python
# 기본 문법
# 변수명 = lambda 매개변수 : 리턴 값
square = lambda x : x**2
print(f"2의 제곱 : {square(2)}")

# 매개변수가 여러 개인 경우
add = lambda a, b: a + b
print(f"10 + 20 = {add(10,20)}")
```

## 고차 함수 map(), filter(), sorted()
 - 고차 함수란 함수를 인자로 받거나 함수를 반환하는 함수를 말한다.

```python
# map : 모든 요소에 함수를 적용하여 새로운 형태를 만듦 , 이더러블한 타입에게 특정 기준으로 가공
nums = [1,2,3,4,5]
squared_nums = list(map(lambda x : x ** 2, nums)) # 가공기준
print(f"제곱 결과 : {squared_nums}")

# filter : 조건이 True인 요소만 골라냄 , 이더러블한 타입에게 특정 기준으로 필터링
evens = list(filter(lambda x : x % 2 == 0 , nums)) # 필터링 기준
print(f"짝수 필터링 : {evens}")

# sorted : key 인자는 "각 요소를 무엇으로 변환해서 비교할 것인가"를 결정하는 함수를 받음 , 이더러블한 타입에게 특정 기준으로 정렬
# 학생 명단 : (이름, 수학점수, 영어점수)
students = [
    ("허준", 80, 90),
    ("이순신", 100, 70),
    ("신사임당", 90, 95)
]

# 수학 점수 기준 정렬
by_math = sorted(students, key=lambda x : x[1])
print(f"수학 점수 기준 정렬 : {by_math}")

# 영어 점수 기준 정렬(단, 내림차순)
by_eng = sorted(students, key=lambda x : x[2], reverse=True)
print(f"영어 점수 기준 정렬 : {by_eng}")

# 이름의 길이 기준 정렬
by_name_len = sorted(students, key=lambda x : len(x[0]))
print(f"이름 길이 기준 정렬 : {by_name_len}")
```
