# 튜플
 - 리스트와 유사하게 일련의 값이 모인 집합을 다루기 위한 자료형으로 **순서**가 있음
 - 하지만 한 번 할당 된 내용을 **변경 할 수 없는(Immutable)** 자료구조

## 주요 특징
 - '()' 소괄호를 사용하거나, 괄호 없이 ','로 나열 하여 생성 => 튜플로 자동 인식
 - 요소 수정이 불가능하여 데이터의 **안정성**이 보장

```python
# 튜플 생성
tp1 = (1, 2, 3)
tp2 = 10, 20, 30 # 소괄호가 없어도 콤마로 연결 되면 튜플로 생성(패킹)

print(tp1)
print(tp2)
print(tp1[1])

# 중첩 튜플 생성 (튜플 안의 튜플)
tuples = (1,2,"hello") , ("test",1,2,3)
print(f"전체 튜플 : {tuples}")
print(f"첫번째 요소 : {tuples[0][2]}")
```

# 튜플의 특징 : 불변성
 - 튜플은 요소 값을 한 번 할당하면 추가, 수정, 삭제가 불가능하다.
 - 만약 값을 변경하려고 하면 'Type Error'가 발생한다.

```python
# 수정 시도(에러 발생)
test_tp = (1,2,3)
# test_tp[0] = 100

# 불변 객체와 재할당의 차이
# 값을 바꾸는 것이 아니라, 새로운 변수를 만드는 것임을 주소 값으로 확인
print(f"기존 주소 : {id(test_tp)}")
test_tp = (4,5,6) # => 요소를 수정한 것이 아니라 변수에 새로운 내용을 할당함
print(f"재할당 후 주소 : {id(test_tp)}")
```

# 시퀀스 자료형 (Sequence Type) 공통 규칙
 - **문자열, 리스트, 튜플** 처럼 여러 값을 순서대로 저장하는 데이터 구조를 '시퀀스'라고 한다.
 - 모든 시퀀스는 인덱스를 가지고 있어 아래 기능을 공통적으로 사용할 수 있다.

```python
nation = ('대한민국', '미국', '중국', '일본', '프랑스')

# 인덱스 : 특정 위치의 값 추출
print(f"인덱싱 : {nation[3]}")

# 슬라이싱 : [시작 : 끝 : 간격]
print(f"슬라이싱 : {nation[0:3]}")
print(f"슬라이싱 : {nation[0:5:2]}")

another = ('영국', '독일')
# 연산 : 합치기(+), 반복(*)
print(f"합치기 : {nation + another}")
print(f"합치기 : {nation *2}")

# 포함 여부(in) 및 길이(len)
print(f"포함 여부 : {'미국' in another}")
print(f"길이 : {len(nation)}")
```
