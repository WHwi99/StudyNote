# 숫자형(Numberic)
숫자를 표현하는 자료형으로 정수(int)와 실수(float)가 대표적이다.

```python
# 정수(int) : 소수점이 없는 숫자
num1 = 10
num2 = -5
print(num1, type(num1))
print(num2, type(num2))
```

```python
# 실수(float) : 소수점이 있는 숫자
num3 = 3.14
num4 = 1.0
print(num3, type(num3))
print(num4, type(num4))
```

# 산술 연산
기본적인 사칙연산과 파이썬에서 제공하는 다양한 연산자가 있다.

```python
# 기본 사칙 연산
a = 10
b = 3

print("더히기", a+b)
print("뺴기", a-b)
print("곱하기", a*b)
print("나누기", a/b) #나누기의 결과 값은 항상 실수 값으로 반환 된다.
```

#  활용도가 높은 연산자
몫, 나머지, 거듭제곱 연산 등이 있다. 실제 프로그래밍에서 매우 자주 사용 된다.

```python
money = 11
person = 3

print("몫 : ", money // person)
print("나머지 : ", money % person)
print("거듭제곱 : ", 2**10)

# [참고] 정수와 실수의 혼합 연산 => 결과는 항상 실수
print(10+0.1)
```
